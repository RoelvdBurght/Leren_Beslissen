```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn import metrics
from scipy.spatial.distance import cdist
import numpy as np
import matplotlib.pyplot as plt
import csv
import re

data = pd.read_csv("kmeans_cluster_all.csv", delimiter=',')
X = data.values

data_begeleiding = pd.read_csv("kmeans_cluster_begeleiding.csv", delimiter=',')
data_location_time = pd.read_csv("kmeans_cluster_location_time.csv", delimiter=',')
data_apps = pd.read_csv("kmeans_cluster_apps.csv", delimiter=',')

all_data = data.values
begeleiding = data_begeleiding.values
location_time = data_location_time.values
apps = data_apps.values

print(begeleiding)
```

```python
dict_all = dict()
for i in range(len(all_data)):
    for j in all_data[i]:
        if j not in dict_all:
            dict_all[j] = 1
        else:
            dict_all[j] += 1

dict_begeleiding = dict()
for i in range(len(begeleiding)):
    for j in begeleiding[i]:
        if j not in dict_begeleiding:
            dict_begeleiding[j] = 1
        else:
            dict_begeleiding[j] += 1

dict_location_time = dict()
for i in range(len(location_time)):
    for j in location_time[i]:
        if j not in dict_location_time:
            dict_location_time[j] = 1
        else:
            dict_location_time[j] += 1
            
dict_apps = dict()
for i in range(len(apps)):
    for j in apps[i]:
        if j not in dict_apps:
            dict_apps[j] = 1
        else:
            dict_apps[j] += 1
            
print(dict_all, dict_begeleiding, dict_location_time, dict_apps)
```

```python
import matplotlib.pyplot as plt

plt.bar(range(len(dict_all)), list(dict_all.values()), align='center')
plt.xticks(range(len(dict_all)), list(dict_all.keys()))
plt.title("All questions")
plt.xlabel("Classes")
plt.ylabel("Number of participants")
plt.show()

plt.bar(range(len(dict_begeleiding)), list(dict_begeleiding.values()), align='center')
plt.xticks(range(len(dict_begeleiding)), list(dict_begeleiding.keys()))
plt.title("Begeleiding")
plt.xlabel("Classes")
plt.ylabel("Number of participants")
plt.show()

plt.bar(range(len(dict_location_time)), list(dict_location_time.values()), align='center')
plt.xticks(range(len(dict_location_time)), list(dict_location_time.keys()))
plt.title("Location & time")
plt.xlabel("Classes")
plt.ylabel("Number of participants")
plt.show()

plt.bar(range(len(dict_apps)), list(dict_apps.values()), align='center')
plt.xticks(range(len(dict_apps)), list(dict_apps.keys()))
plt.title("Apps")
plt.xlabel("Classes")
plt.ylabel("Number of participants")
plt.show()
```
