```python
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("one_hot_data.csv")
df
```

```python
labels_age = ['<20', '20-29', '30-39', '40-49', '50-59', '>60']
bins_age = [0, 20, 29, 39, 49, 59, 100]

labels_height = ['<160', '160-169', '170-179', '180-189', '190-199', '>200']
bins_height = [0, 159, 169, 179, 189, 199, 1000]

labels_weight = [ '<60', '60-69', '70-79', '80-89', '90-99', '>100']
bins_weight = [0, 59, 69, 79, 89, 99, 1000]

q48 = pd.cut(df["Q48"], bins_age, labels=labels_age)
q49 = pd.cut(df['Q49'], bins_height, labels=labels_height)
q50 = pd.cut(df['Q50'], bins_weight, labels=labels_weight)
q48 = q48.cat.add_categories([0])
q49 = q49.cat.add_categories([0])
q50 = q50.cat.add_categories([0])

q48.fillna(0)
q49.fillna(0)
q50.fillna(0)
```

```python
# q50.value_counts(sort=False).plot(kind='bar')
q49.value_counts(sort=False).plot(kind='bar')
```

```python
q48.value_counts(sort=False).plot(kind='bar')
```

```python
df['Q48'] = q48
df['Q49'] = q49
df['Q50'] = q50
df
```

```python
final_df = pd.get_dummies(df)
final_df.to_csv("final_data.csv")
```
