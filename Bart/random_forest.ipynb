```python
print(__doc__)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from kmodes.kmodes import KModes
from sklearn.cluster import KMeans

from sklearn.datasets import make_classification
from sklearn.ensemble import ExtraTreesClassifier, RandomForestClassifier
pd.set_option('display.max_column', None)
from sklearn import tree

```

```python
data = pd.read_csv("../Data/final_data.csv", delimiter=',')

drop = [i for i in range(57,167)] + [i for i in range(35,47)]
data = data.drop(data.columns[drop], axis=1)
X = data.values
data
```

```python
kmeans = KMeans(n_clusters=4, n_init=10, verbose=1).fit(X)
labels = kmeans.labels_
```

```python
forest = ExtraTreesClassifier(n_estimators=250,
                              random_state=0)
```

```python
forest.fit(X, labels)
importances = forest.feature_importances_
std = np.std([tree.feature_importances_ for tree in forest.estimators_],
             axis=0)
indices = np.argsort(importances)[::-1]

# Print the feature ranking
print("Feature ranking:")

for f in range(X.shape[1]):
    print("%d. feature %d (%f)" % (f + 1, indices[f], importances[indices[f]]))
```

```python
# print(data.columns[indices])
# Plot the feature importances of the forest
plt.figure()
plt.title("Feature importances")
print()
print(X.shape)
first = importances[0:7]
print(first)
plt.bar(range(X.shape[1]-38), first,
       color="r",  align="center")
plt.xticks(range(X.shape[1]), indices)
plt.xlim([-1, X.shape[1]-38])
plt.show()
```

```python
# Plot the feature importances of the forest
first_i = 30
names = []
for i in indices[:first_i]:
    names.append(data.columns[i])
plt.figure()
plt.title("Feature importances")
print(importances[indices])
print(importances)
plt.bar(range(X.shape[1]-(X.shape[1]-first_i)), importances[indices][:first_i],
       color="r", align="center")
plt.xticks(range(X.shape[1]), names, rotation='vertical')
plt.xlim([-1, X.shape[1]-(X.shape[1]-first_i)])
plt.show()
```
